## Adaptación en PySal

```python
import pandas as pd
import numpy as np
import pysal as ps
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
agebs = gpd.read_file('data/agebs_cdmx.shp')
agebs = agebs.set_index('ageb_urban')
agebs.head()
```

```python
agebs['tmp']
```

```python
m_reina = ps.lib.weights.Queen.from_shapefile('data/agebs_cdmx.shp', idVariable = 'ageb_urban')
m_reina
```

```python
import pejelagarto
```

```python
z = pejelagarto.Random(agebs['tmp'] , m_reina , 5)
```

```python
z.output
```

```python
agebs['clusterpy_A'] = z.output['r2a']
n = agebs[['clusterpy_A' , 'geometry']]
n = n.dissolve(by = 'clusterpy_A')
n.plot(figsize = (10,10))
```

```python
z.w
```

```python
y = agebs['tmp']
dictionary = { i : y.index[i] for i in range(0, len(y.index) ) }
dictionary_r = {v: k for k, v in dictionary.items()}

y = y.reset_index(drop = True)

y = y.transform(lambda x : [x])
y = y.to_dict()
y
```

```python
dictionary_r
```

```python
w = m_reina.neighbors
for i in w:
    vecinos = w[i]
    vecinos = [dictionary_r.get(item,item) for item in vecinos]
    w[i] = vecinos

w = dict((dictionary_r[key], value) for (key, value) in w.items())
w

#tmp = pd.Series(m_reina.neighbors)
#tmp = tmp.replace({'090150001056A':0})
#tmp.to_dict()
```
